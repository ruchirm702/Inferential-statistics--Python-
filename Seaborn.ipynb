# Seaborn samples

```python
import seaborn as sns
sns.get_dataset_names()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
df = sns.load_dataset('planets')
df.head()
```

```python
sns.set_style('darkgrid')
sns.displot(df['mass'],kde=False)
plt.show()
```

```python
sns.set_style('darkgrid')
sns.distplot(df['mass'])
plt.show()
```

```python
df =  sns.load_dataset('iris')
df.head()
sns.jointplot(x='petal_length',y='sepal_length',data=df)
plt.show()
```

```python
df =  sns.load_dataset('iris')
df.head()
sns.jointplot(x='petal_length',y='sepal_length',data=df,kind='hex')
plt.show()
```

```python
sns.set_style("ticks")
sns.pairplot(df,hue='species',diag_kind="kde",kind="scatter",palette="Greens")
```
