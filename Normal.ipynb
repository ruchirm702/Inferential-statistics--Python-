# Normal Distribution

```python
"""
Average Height 167cms
Standard Dev 20 cms
"""
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
mu = 167
sigma = 20

height_samp = np.random.normal(loc=mu,scale=sigma,size=10000)
height_samp[:10]
```

```python
plt.hist(height_samp,bins=50)
```

```python
heightRequired = 5*12*2.54
height_dist = stats.norm(loc=mu,scale=sigma)
height_dist.cdf(heightRequired)
```

```python
heightRequired2 = (6*12*2.54)+(5*2.54)
1-height_dist.cdf(heightRequired2)
```

```python
ax_r =  np.arange(100,240,0.1)
plt.plot(ax_r,height_dist.pdf(ax_r))
```

```python
plt.plot(ax_r,height_dist.pdf(ax_r))
section = range(int(heightRequired),int(heightRequired2)+1)
plt.fill_between(section,height_dist.pdf(section),color='lightblue')
height_dist.cdf(int(heightRequired2)) - height_dist.cdf(int(heightRequired))
```

```python
height_dist.ppf(0.975)
```
